# First Try Yolo

```python
import os
from dotenv import load_dotenv
from ultralytics import YOLO

load_dotenv()

ENV_TYPE = os.getenv("ENV_TYPE", "local")
dataset_path = "" if ENV_TYPE=="gcp" else os.getenv("LOCAL_DATA_PATH")
```

```python
model = YOLO("yolov8n.pt")
model.train(
    data="./yolo/data.yaml",
    epochs=1,
    imgsz=640,
    batch=10
)
```

```python
metrics = model.val()

results = model.predict("../data/cow-hooves")
print(results)
```
